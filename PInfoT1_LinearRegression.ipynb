```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprice.csv')
df.head()
```

```python
display(df)
```

```python
X = df[['Area', 'Bedroom', 'Bathroom']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
lr  = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print("Accuracy Score : ", lr.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error

print("MSE : ", mean_squared_error(y_test, lr.predict(X_test)))
```

```python
plt.title("Actual vs Predicted Graph")
plt.scatter(y_test,lr.predict(X_test), color = 'blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.ylabel("Predicted Price")
plt.xlabel("Actual Price")
```

```python
area = int(input("Enter Area of the House : "))
bedroom = int(input("Enter Number of Bedroom : "))
bathroom = int(input("Enter Number of Bathroom : "))

p = lr.predict([[area,bedroom,bathroom]])

print("Price For the house Will May be : ", int(p[0]))
```
